# Model Validation( Thẩm định model )

* There are many metrics for summarizing model quality, but we'll start with one called **Mean Absolute Error(Lỗi tuyệt đối trung bình)** (also called MAE). Let's break down this metric starting with the last word, error.
* The prediction error for each house is:

      error=actual−predicted

&ensp;&ensp;&ensp;&ensp;&ensp;So, if a house cost 150,000 and you predict it would cost 100,000 the error is $50,000.

## Calculate the MAE - Mean Absolute Error

```
from sklearn.metrics import mean_absolute_error
predicted_home_price = home_model.predict(X)
mean_absolute_error(y, predicted_home_price)
```

Scikit-learn có công cụ giúp tách traning set và test set đó là hàm train_test_split.
Nguyên nhân tại sao ta phải áp dụng việc phân chia train set và test set.
* Đơn giản có thể hiểu rằng việc ta kiểm tra model dựa trên 1 loại dữ liệu đã được đào tạo sẽ có tính chính xác cao. Tuy nhiên kết quả không thể đảm bảo với dữ liệu chưa nằm trong huấn luyện.
* Đó là lý do cho việc ta phân chia tập dữ liệu thành train_set và test_set.
* Ta có thể trích 1 phần dữ liệu trong train_set thành validation_set để kiểm tra model với unseen data.

```
from sklearn.model_selection import train_test_split
# random_state giúp đảm bảo việc phân chia train_Set và test_set là như nhau tại mọi thời điểm. 
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
# val_x = validation X.
home_model = DecisionTreeRegressor()
home_model.fit(train_X,train_y)
val_prediction = home_model.predict(val_X)
print(mean_absolute_error(val_y,val_prediction))
```

Ta thấy rõ sự khác biệt khi ta kiểm tra trên bộ dữ liệu mới tách

* Đối với dữ liệu đã huấn luyện: MAE = 62,35
* Đối với dữ liệu không qua huấn luyện = 32953.56

=> Thực tế model này không có tính khả thi khi áp dụng vào thực tế.

# Underfitting and Overfitting

Đây là vấn đề thường xảy ra khi xử lý model.

Như ở phần trên ta có thể thấy model hoạt động cực kì hiệu quả với train_set nhưng lại gặp vấn đề với test_set -> có thể hiểu trường hợp này là overfitting.

Nếu model hoạt động không hiệu quả ở cả train_set và test_set -> được gọi là underfitting.

```
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 25, 50, 100, 250, 500]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

Có thể thấy 50 chính là số lá nhiều nhất tối ưu.

Hay nói cách khác 50 lá chính là fit model.

## Tinh chỉnh model.

```
# Fit the model with best_tree_size. Fill in argument to make optimal size
final_model = DecisionTreeRegressor(max_leaf_nodes=100, random_state=1)

# fit the final model
final_model.fit(X, y)
```

```
val_prediction = final_model.predict(val_X)
print(mean_absolute_error(val_y,val_prediction))
```

Ta có thể thấy giá trị MAE đã giảm rõ rệt so với trước khi xử lý overfitting.
